#Gradient Descent


```python
new_x=3   #starting point for gradient descent- random guess to the.
previous_x=0 # ponto onde estamos
step_multiplier=0.1   #learning rate
precision=0.0001

#LISTS MONITORAR VALORES
x_list=[new_x]
slope_list=[df(new_x)]

#First step of ML. PREDICT
#Step 2- calculate the error  - DERIVADA
for n in range(500):
  previous_x=new_x
  gradient=df(previous_x)  #achar a derivada do ponto que estamos. Mais longe do ponto de menor custo, maior a derivada.. maior erro; mais perto, menor erro
  new_x= previous_x - step_multiplier * gradient  #Descer a funÃ§Ã£o  (menos -) o step proporcional ao gradient da funÃ§Ã£o onde estamos. Subtrair do previous_x, porque queremos ANDAR CONTRA O GRADIENTE, PARA O MÃNIMO.Andar na funÃ§Ã£o

  #new_x and previous_x are GETTING CLOSER
  step_size=   abs(new_x - previous_x)
  x_list.append(new_x)
  slope_list.append(df(new_x))

  #print(step_size)
  if step_size < precision:
    print(f"Loop ran this many times {n}")
    break



print('Local minium occurs at:', new_x )
print('Slope or df(x) value at this points is:', df(new_x) ) #Valor da derivada no ponto
print('f(x) value or cost at this point', f(new_x))

#inrange variando, derivada muda. Descendo o valley, cost function, atÃ© chegar no menor valor

#new_x and previous_x are GETTING CLOSER . DIFERENÃA ENTRE ELES Ã QUASE 0



```

#Gradient Descent as a Python Function

```python

def gradient_descent(derivative_func,initial_guess,multiplier=0.02,precision=0.001,max_iter=300):
  new_x=initial_guess
  x_list=[new_x]
  slope_list=[derivative_func(new_x)]

  for n in range(max_iter):
    previous_x=new_x
    gradient=derivative_func(previous_x)
    new_x= previous_x - multiplier * gradient

    step_size=abs(new_x - previous_x)
    x_list.append(new_x)
    slope_list.append(derivative_func(new_x))


    if step_size < precision:
      break

  return new_x,x_list,slope_list
```

```python
#Call the function . 3 values will be returned. store- variables

#Se chamar de 'data_tuple' o gradient descent, posso chamar 'data_tuple[0]' pro primeiro return da funÃ§Ã£o

local_min,list_x,deriv_list = gradient_descent(dg,0.5,0.02,0.001)
print('Local min occurs at', local_min)
print('Number of steps', len(list_x))



```

## Batch Gradient Descent with SymPy

```python
#Setup
multiplier = 0.1
max_itter=200
params=np.array([1.8,1.0]) #initial guess

for n in range(max_itter):
  gradient_x=diff(f(a,b),a).evalf(subs={a:params[0],b:params[1]})
  gradient_y=diff(f(a,b),b).evalf(subs={a:params[0],b:params[1]})
  #WORK WITH ARRAYS
  gradients=np.array([gradient_x,gradient_y])
  #O quÃ£o longe estamos do menor custo, combase na derivada nos pontos
  params=params-multiplier*gradients  #current value of our parameters subtracted multiplier*gradient, replacing on 'params'



#Results
print('Values in gradient array', gradients)
print('Minimum occurs at x value of:', params[0])
print('Minimum occurs at y value of:', params[1])
print('Cost is:',f(params[0],params[1]))




```

## $$\frac{\partial f}{\partial x}= \frac{2x\ln(3)\cdot3^{-x^2-y^2}}{3^{-x^2-y^2}+1}$$

```python
#MAIS PERFORMÃTICO nÃ£o usar SYMPY e usar a funÃ§Ã£o no python

def fpx(x,y):
  r=3**(-x**2- y**2)
  return 2*x*log(3)*r/(r+1)**2

def fpy(x,y):
  r=3**(-x**2- y**2)
  return 2*y*log(3)*r/(r+1)**2
```

```python
fpx(1.8,1.0)
```

```python
#Setup
multiplier = 0.1
max_itter=500
params=np.array([1.8,1.0]) #initial guess

for n in range(max_itter):
  gradient_x=fpx(params[0],params[1])
  gradient_y=fpy(params[0],params[1])
  #WORK WITH ARRAYS
  gradients=np.array([gradient_x,gradient_y])
  #O quÃ£o longe estamos do menor custo, combase na derivada nos pontos
  params=params-multiplier*gradients  #current value of our parameters subtracted multiplier*gradient, replacing on 'params'



#Results
print('Values in gradient array', gradients)
print('Minimum occurs at x value of:', params[0])
print('Minimum occurs at y value of:', params[1])
print('Cost is:',f(params[0],params[1]))




```

##Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = -\frac{2}{n} \sum_{i=1}^{n} \big(y^{(i)} - \theta_0 - \theta_1 x^1{(i)} \big)$$


## $$\frac{\partial MSE}{\partial \theta_1} = -\frac{2}{n} \sum_{i=1}^{n} \big(y^{(i)} - \theta_0 - \theta_1 x^1{(i)} \big) \big(x^{(i)})$$

##MSE & Gradient Descent

```python
#x values, y values, array of THETAS (theta 0 index 0, theta 1 index 1)
#derivadas parciais da funÃ§Ã£o de custo

def grad(x,y,thetas):
    n=y.size


    theta0_slope= (-2/n) * sum(y-thetas[0]-thetas[1]*x)
    theta1_slope=(-2/n) * sum((y-thetas[0]-thetas[1]*x)*x)

    #return np.array([theta0_slope[0],theta1_slope[0]])
    #return np.append(arr=theta0_slope,values=theta1_slope)
    return np.concatenate((theta0_slope,theta1_slope),axis=0)





```

```python
#Run gradient descent algorythm

multiplier= 0.01
thetas=np.array([2.9,2.9])  #Initial guesses

#Collect data to scatter plot
plot_vals=thetas.reshape(1,2)
mse_vals=mse(y_5,thetas[0]+thetas[1]*x_5)



for i in range(1000):
    thetas=thetas-multiplier*grad(x_5,y_5,thetas)  #thetas vÃ£o sendo atualizados

    #Append the new values to our numpy arrays
    plot_vals=np.concatenate((plot_vals,thetas.reshape(1,2)),axis=0) # Concatenar no array plot vals, os novos valores gerados de thetas
    mse_vals=np.append(arr=mse_vals,values=mse(y_5,thetas[0]+thetas[1]*x_5))  # y_5 vs y_hat predicted values


#Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta1:', thetas[1])
print('MSE is:', mse(y_5,thetas[0]+thetas[1]*x_5))  # y_5 vs y hat

#CONVERGIU NESSES VALORES QUE MINIMIZAM A (NOSSA FUNÃÃO DE CUSTO/o erro quadrÃ¡tico da mÃ©dia)

#Precisamos rastrear os valores que o loop estÃ¡ gerando

plot_vals

```
